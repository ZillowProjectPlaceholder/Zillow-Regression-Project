Some of the columns could have better names.

```python
df = df.rename(columns={'bathroomcnt': 'bathroom', 'bedroomcnt': 'bedroom', 'calculatedfinishedsquarefeet': 'sqft',
                        'taxvaluedollarcnt': 'propertytaxvalue', 'lotsizesquarefeet': 'lotsqft'})
```

Pool count has a minimum of 1 and a lot of null values so it is probably safe to assume that a null is a house with no pool. We will replace the nulls with zeros.

```python
df['poolcnt'] = df.poolcnt.fillna(0)
```

```python
df.head()
```

There are properties with a room count of zero but they have bedrooms and bathrooms. It being recorded as a zero is most likely an error and so I will be replacing any zero in roomcnt with the minimum number of rooms recorded, being bedroom + bathroom.

```python
df['roomcnt'] = np.where(df['roomcnt'] == 0.0, (df.bathroom + df.bedroom), df['roomcnt'])
```

```python
df.head()
```

```python
df[df['fullbathcnt'].isnull()]
```

Because there are only 46 rows where fullbathcnt is null, and these all have their own problems with incomplete data we will be dropping rows where fullbathcnt is null.

```python
df = df[~df['fullbathcnt'].isna()]
```

```python
df['fullbathcnt'].isna().sum()
```

```python
(df['bedroom'] == 0).sum() + (df['bathroom'] == 0).sum()
```

There are still 6 houses with no bedroom and bathrooms recorded so we will be dropping these as well as they are very rare cases that are either errors, incomplete, or outliers.

```python
df = df[df['bedroom'] != 0]
df.shape
```

```python
df.info()
```
